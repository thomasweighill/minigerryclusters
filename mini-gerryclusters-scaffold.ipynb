# K-means on plans

This is a template with some information on how to run K-means using `minigerryclusters`.

## Input format

The K-means algorithm takes in an ensemble which is a numpy array of shape
`(number of plans) x (number of districts per plan) x (number of precincts)`

## Output format

The output of the $K$-means algorithm contains:
- `means`: the means as vectors of length $p$
- `indexings`: for each plan $P$, a vector $\phi$ of indexings so that the $i^{th}$ district in the mean it belongs to matches to the $\phi_i^{th}$ district in $P$.
- `matching`: a list containing the cluster assignment for each plan
